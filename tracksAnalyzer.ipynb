```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud


# Load the data
data = pd.read_csv('tamil_tracks.csv')


# Load the dataset
df = data

```

```python
# List of numerical columns to plot histograms
numerical_columns = data.select_dtypes(include=['float64', 'int64']).columns

# Create and save a histogram for each numerical attribute
for column in numerical_columns:
    plt.figure(figsize=(4, 2))
    plt.hist(data[column].dropna(), bins=30, alpha=0.75, color='r')
    plt.title(f'Histogram of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.grid(False)
    plt.show()
    # plt.savefig(f'./graphs/histogram_of_{column}.png')  # Saves the plot as a PNG file
    plt.close()  # Close the plot to free memory

```

```python
# Count the number of tracks per language
language_counts = data['language'].value_counts()

# Define a color palette
colors = ['red', 'yellow', 'orange', 'green', 'blue', 'purple', 'cyan']  # Adjust more colors as needed

# Create a pie chart
plt.figure(figsize=(10, 10))
plt.pie(language_counts, labels=language_counts.index, autopct='%1.1f%%', startangle=140, colors=colors[:len(language_counts)])
plt.title('Distribution of Tracks by Language')
plt.show()
```

```python

# Split artist names and flatten the list
all_artists = data['artist_name'].str.split(',').explode()

# Remove any leading/trailing whitespace
all_artists = all_artists.str.strip()

# Count the number of tracks per artist
artist_counts = all_artists.value_counts()

# Filter to show only artists with more than a certain number of tracks for better visualization
filtered_artist_counts = artist_counts[artist_counts > 1000]  # Adjust this threshold as needed

# Create a bar plot for artist distribution
plt.figure(figsize=(12, 8))
filtered_artist_counts.plot(kind='bar', color='teal')
plt.title('Number of Tracks per Artist (Filtered for More than 1000 Tracks)')
plt.xlabel('Artist')
plt.ylabel('Number of Tracks')
plt.xticks(rotation=90)  # Rotate labels for better readability
plt.show()
```

```python
# Filter to show only artists with more than a certain number of tracks for better visualization
filtered_artist_counts = artist_counts[artist_counts > 800]  # Adjust this threshold as needed

# Create a bar plot for artist distribution
plt.figure(figsize=(12, 8))
filtered_artist_counts.plot(kind='bar', color='blue')
plt.title('Number of Tracks per Artist (Filtered for More than 800 Tracks)')
plt.xlabel('Artist')
plt.ylabel('Number of Tracks')
plt.xticks(rotation=90)  # Rotate labels for better readability
plt.show()
```

```python
# Filter for English tracks
english_data = data[data['language'] == 'English']

# Split artist names and flatten the list
all_artists = english_data['artist_name'].str.split(',').explode()

# Remove any leading/trailing whitespace
all_artists = all_artists.str.strip()

# Combine all artist names into one large string
artist_text = ' '.join(all_artists)

# Generate the word cloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(artist_text)

# Display the word cloud using matplotlib
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')  # Turn off the axis
plt.title('Word Cloud of Artists from English Tracks')
plt.show()
```

```python
# Group data by year and calculate the mean popularity
yearly_data = data.groupby('year')['popularity'].mean().reset_index()

# Setting up the plot
plt.figure(figsize=(10, 6))

# Creating a line plot
sns.lineplot(x='year', y='popularity', data=yearly_data, marker='o')

# Adding titles and labels
plt.title('Average Popularity of Tracks Over the Years')
plt.xlabel('Year')
plt.ylabel('Average Popularity')
plt.grid(True)

# Show the plot
plt.show()
```
